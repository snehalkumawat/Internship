# 9.  Write a python program to scrape fresher job listings from ‘https://internshala.com/’.
    It should include job title, company name, CTC, and apply date.

```python
from bs4 import BeautifulSoup
import requests
page1=requests.get("https://internshala.com/fresher-jobs")
page1
```

```python
Soup=BeautifulSoup(page1.content)
Soup
```

```python
job_title=Soup.find_all("div",class_="heading_4_5 profile")
job_title
```

```python
Job_title=[]  #empty list

for i in job_title:
    for j in i.find_all("a"):
        Job_title.append(j.text.replace("\n",""))
    
Job_title
```

```python
company_name=Soup.find_all("div",class_="heading_6 company_name")
company_name
```

```python
Company_name=[]  #empty list

for i in company_name:
    for j in i.find_all("a"):
        Company_name.append(j.text.replace("\n",""))
Company_name
```

```python
ctc=Soup.find_all("div",class_="item_body")
ctc
```

```python
CTC=[]  #empty list

for i in ctc:
    CTC.append(i.text.replace("\n",","))
CTC=CTC[1::3]
CTC
```

```python
Apply_date=[]  #empty list

for i in ctc:
    Apply_date.append(i.text.replace("\n",","))
#Apply_date=Apply_date[::]    
Apply_date=Apply_date[2::3]
Apply_date
```

```python
import pandas as pd
Fresher_job_listings=pd.DataFrame({})
Fresher_job_listings['Job_title']=Job_title
Fresher_job_listings['Company_name']=Company_name
Fresher_job_listings['CTC']=CTC
Fresher_job_listings['Apply_date']=Apply_date
Fresher_job_listings
```
